## 线性回归

$$ L(w,b) = \frac{1}{n} \sum_{i\in[n]} (y_i - (w^Tx_i + b))^2 $$

你可能会用到的函数有：`numpy.linalg.inv()`、`numpy.dot()`、`numpy.transpose()`、`numpy.ones()`、`numpy.concatenate()`、`numpy.sum()` 等。
